# Check out contour plot

````python
import numpy as np
import matplotlib.pyplot as plt

def plot_countour(depth = 2):
   
    iris = load_iris()
    X = iris.data[:, 2:] # petal length and width
    y = iris.target

    tree_clf = DecisionTreeClassifier(max_depth=depth)
    tree_clf.fit(X, y)

    n_classes = 3
    plot_colors = "ryb"
    plot_step = 0.02

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    #plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = tree_clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.5)

    plt.xlabel(iris.feature_names[2])
    plt.ylabel(iris.feature_names[3])

    # Plot the training points
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=color, label=iris.target_names[i],
                    cmap=plt.cm.RdYlBu, edgecolor='black', s=20)


    plt.suptitle("Decision surface of a decision tree using paired features")
    plt.legend(loc='lower right')
````

````python
plot_countour(depth=2)
````

# Impureza Gini

Mede a impureza de um determinado atributo.

```
se gini (nó) = 0, o nó é puro
se gini (nó) próximo a 1, então o nó é impuro
```

\begin{equation}G_i = 1 - \sum_{k =1}^{n}p_{i,k}^2\end{equation}

Onde:
* $G_1$ é o fator de impureza ginni
* $p_{i, k}$ é a proporção de instâncias da classe k entre as instâncias de treinamento $i$-ésimo nó.

# Algoritmo de treinamento CART

* O algoritmo divide primeiro o conjunto de treinamento em dois subconjuntos, usando um único recurso $ k $ e um limite $ t_k $
* Procura o par $ (k, t_k) $ que produz o subconjunto mais puro usando a métrica Gini.

Função de custo

$$J\left(k, t_{k}\right)=\frac{m_{\mathrm{left}}}{m} G_{\mathrm{left}}+\frac{m_{\mathrm{right}}}{m} G_{\mathrm{right}}$$

Onde:

* $G_{\mathrm{left/right}}$ mede a impureza dos subconjuntos.
* $m_{\mathrm{left/right}}$ é o número de instâncias de subconjuntos.
* $m$ número total de amostras

# Implementando a função de custo do CART

````python
def split_sets(dataset, feature, threshold):
    left = dataset[dataset[feature] <= threshold]
    right = dataset[dataset[feature] > threshold]
    
    return left, right
````

# Implementando a Função Gini

````python
def gini(classes, nodeset):
    counts = nodeset.value_counts()
    
    proportions = []
    for label in classes:
        if label not in counts:
            proportions.append(0)
        else:
            proportions.append(counts[label]/len(nodeset))
    
    G = 1 - np.sum([p**2 for p in proportions])
    return G
````

````python
left, right = split_sets(iris_df, "petal_length", 2.45)
gini(classes, left.species)
````

# Implementando a Função Gini

````python
def calculate_cost(dataset, feature, threshold, target_name, classes):
    
    left_set, right_set = split_sets(
        dataset, feature, threshold
    )
    
    gini_left = gini(classes, left_set[target_name])
    gini_right = gini(classes, right_set[target_name])
    
    total_len = len(left_set) + len(right_set)
    
    J = gini_left*len(left_set)/total_len + gini_right*len(right_set)/total_len
    return J
````

````python
calculate_cost(iris_df, "petal_length", 2.4, "species", classes)
````

# Check Implementation

````python
dataset = iris_df
feature = "petal_length"
threshold = 2.4
target_name = "species"

feature_min = min(dataset[feature])
feature_max = max(dataset[feature])

featrange = np.arange(feature_min, feature_max, 0.1)

costs = [
    calculate_cost(dataset, feature, threshold, "species", classes)
    for threshold in featrange
]

plt.plot(featrange, costs)
plt.title(f"Cost for feature {feature}")
plt.xlabel("Classes")
plt.ylabel("Costs")
````
